```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
idc = pd.read_csv('IDSCountry.csv')
idc = pd.DataFrame(idc)
idc.head()
```

```python
idc.tail()
```

```python
idc.shape
```

```python
idc.info()
```

```python
idc.isnull().sum()
```

```python
idc = idc.replace(np.nan," ",regex=True)
```

```python
idc.isna().sum()
```

```python
idc.dtypes
```

```python
idc['National accounts reference year']=idc['National accounts reference year'].str.replace(" ","0")
```

```python
idc=idc.astype({'National accounts reference year':'float64','Latest industrial data':'str','Latest trade data':'str','Latest water withdrawal data':'str'})
```

```python
idc.dtypes
```

```python
idc['Latest industrial data']=idc['Latest industrial data'].apply(lambda x :x.split(".")[0])
idc['Latest trade data']=idc['Latest trade data'].apply(lambda x :x.split(".")[0])
idc['Latest water withdrawal data']=idc['Latest water withdrawal data'].apply(lambda x :x.split(".")[0])
```

```python
idc['Latest industrial data']=pd.to_datetime(idc['Latest industrial data'],errors='coerce')
idc['Latest trade data']=pd.to_datetime(idc['Latest trade data'],errors='coerce')
idc['Latest water withdrawal data']=pd.to_datetime(idc['Latest water withdrawal data'],errors='coerce')
```

```python
idc.dtypes
```

```python
idc.drop(['Unnamed: 31'],axis=1,inplace=True)
```

```python
idc.to_csv('IDSCountry_F.csv',index=False,header=True)
```
